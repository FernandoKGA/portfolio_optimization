```python
df['new_date'] = pd.to_datetime(df['DATPRG'].astype(str), format='%Y%m%d')
```

```python
df['CODNEG'] = df['CODNEG'].astype(str)
```

```python
#CRIA DATAFRAME E COLOCA CADA DIA DO ANO NELE (DIAS QUE ESTAO NO DATAFRAME ORIGINAL)
acoes = df['CODNEG'].unique()
dias = df['new_date'].unique()
d = {'data': dias}
df_est = pd.DataFrame(dias, columns=['data'])
```

```python
df_est
```

```python
df['CODNEG'] = df['CODNEG'].str.replace(' ', '')
```

```python
#CRIA COLUNA PARA ACAO E INICIALIZA COM 0
for acao in acoes:
    df_est[acao] = 0 
```

```python
df_est.columns = df_est.columns.str.replace(' ', '')
```

```python
df_est
```

```python
# ESSA FUNCAO IRA PREENCHER OS PRECOS DAS ACOES EM CADA DIA
for acao in acoes:
    
    df_acao = df[df['CODNEG']==acao]
    
    for ind,row in df_acao.iterrows():
        dia = row['new_date']
        match = df_est.loc[ : ,'data'] == dia
        indice = match[match].index
        df_est.iloc[indice, df_est.columns.get_loc(acao)] = row['PREULT']
```
